# ---------------------------------------------- Movie ---------------------------------------------------------

```python
actor=pd.read_csv('actor.csv')

cast=pd.read_csv('cast.csv')

director=pd.read_csv('director.csv')

genres=pd.read_csv('genres.csv')

mov_direct=pd.read_csv('movie_direction.csv')

mov_genre=pd.read_csv('movie_genres.csv')

movie=pd.read_csv('movie.csv')

ratings=pd.read_csv('ratings.csv')

review=pd.read_csv('reviewer.csv')
```

#### Write a SQL query to find those movies, which were released before 1998. Return movie title.

```python
movie[movie['mov_dt_rel'].dt.year<1998][['mov_title']]
```

#### Set the language of movie language as 'Chinese' for the movie which has its existing language as Japanese and the movie year was 2001.

```python
movie.loc[(movie['mov_lang']=='Japanese') & (movie['mov_year']==2001),'mov_lang']='chinese'

movie[movie['mov_year']==2001]
```

#### Write a SQL query to find name of all the reviewers who rated the movie 'Slumdog Millionaire'.

```python
ratings.merge(review,on='rev_id',how='inner').merge(movie[movie['mov_title']=='Slumdog Millionaire'],on='mov_id',how='inner')[['rev_name','mov_title']]
```

#### Write a query which fetch the first name, last name & role played by the actor where output should all exclude Male actors.

```python
cast.merge(actor[actor['act_gender']=='M'],on='act_id',how='inner')[['act_fname','act_lname','role']]
```

#### Write a SQL query to find the actors who played a role in the movie 'Annie Hall'. Fetch all the fields of actor table. (Hint: Use the IN operator). 

```python
actor.merge(cast,on='act_id',how='inner').merge(movie[movie['mov_title']=='Annie Hall'],on='mov_id',how='inner')[['act_id', 'act_fname', 'act_lname', 'act_gender']]
```

#### Write a SQL query to find those movies that have been released in countries other than the United Kingdom. Return movie title, movie year, movie time, and date of release, releasing country.

```python
movie[movie['mov_rel_country']!='UK'][[ 'mov_title', 'mov_year', 'mov_time',  'mov_dt_rel', 'mov_rel_country']] 
```

#### Print genre title, maximum movie duration and the count the number of movies in each genre. (HINT: By using inner join)

```python
genres.merge(mov_genre,on='gen_id',how='inner').merge(movie,on='mov_id',how='inner').groupby('gen_title').agg({'mov_time':'max','mov_id':'count'})
```

#### Create a view which should contain the first name, last name, title of the movie & role played by particular actor. 

```python
actor.merge(cast,on='act_id',how='inner').merge(movie,on='mov_id',how='inner')[['act_fname','act_lname','mov_title','role']]
```

#### Write a SQL query to find the movies with the lowest ratings 

```python
movie.merge(ratings,on='mov_id',how='inner').sort_values('rev_stars',ascending=True).head(1)
```
